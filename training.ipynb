```python
import pandas as pd

# Cargar el archivo Parquet
df = pd.read_parquet("df_restaurants_cleaned_with_sentiment.parquet")
```

```python
# Seleccionar 100 user_id únicos
unique_user_ids = df['user_id'].drop_duplicates().head(100)

# Filtrar el dataset para incluir solo estos 100 user_id
filtered_df = df[df['user_id'].isin(unique_user_ids)]
```

```python
# Guardar el nuevo dataset como parquet
filtered_df.to_parquet("filtered_dataset.parquet", index=False)

```
